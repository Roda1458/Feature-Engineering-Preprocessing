Feature Engineering/Preprocessing

Mean and Variance

```python
import numpy as np

# Example 1xN vector
N = 5
vector = np.array([2, 4, 6, 8, 10])
print("Vector:", vector)

# Compute mean
mean = np.mean(vector)
print("Mean:", mean)

# Compute sample variance
variance = np.var(vector, ddof=1)  # ddof=1 for sample variance
print("Variance (Sample):", variance)
```

Covariance between two vectors represents.

```python
import numpy as np

# Example 1xM vectors
M = 5
vector1 = np.array([2, 4, 6, 8, 10])
vector2 = np.array([1, 3, 5, 7, 9])

# Compute covariance
covariance = np.cov(vector1, vector2)[0, 1]  # Covariance between vector1 and vector2
print("Covariance between vector1 and vector2:", covariance)
```

Correlation

```python
import numpy as np

# Example 1xN vectors
N = 5
vector1 = np.array([2, 4, 6, 8, 10])
vector2 = np.array([1, 3, 5, 7, 9])

# Compute correlation
correlation = np.corrcoef(vector1, vector2)[0, 1]  # Correlation between vector1 and vector2
print("Correlation between vector1 and vector2:", correlation)
```

covariance matrix and correlation matrix.

```python
import numpy as np

# Example MxN feature matrix
M = 3
N = 5

# Random MxN feature matrix
X = np.random.rand(M, N)
print("Feature Matrix X:\n", X)

# Compute the covariance matrix
cov_matrix = np.cov(X)
print("\nCovariance Matrix:\n", cov_matrix)

# Compute the correlation matrix
corr_matrix = np.corrcoef(X)
print("\nCorrelation Matrix:\n", corr_matrix)
```
